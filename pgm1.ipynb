```python
import matplotlib.pyplot as plt 
hours_studied = [10, 9, 2, 15, 10, 16, 11, 16] 
scores = [95, 80, 10, 50, 45, 98, 38, 93] 

plt.figure(figsize=(10, 6)) 
plt.plot(hours_studied, scores, 'r*', linestyle='-', markersize=10)  
plt.xlabel('Number of Hours Studied') 
plt.ylabel('Score in Final Exam') 
plt.title('Effect of Hours Studied on Final Exam Score')
plt.grid(True) 
plt.show() 
```
